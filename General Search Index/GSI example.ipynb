```python
import GSI as gsi
import matplotlib.pyplot as plt
```

# **Examples of Search Index** #

### 1. Canada House Search Index ### 

**"house for sale"** is the initial search term and the search index is the based of search terms related to the initial search "house for sale"




```python
HSI = gsi.search_index_tinator('house for sale',['CA'],1)
```

Plot of the search terms related to buying houses

```python
with plt.style.context('Solarize_Light2'):
    plt.rcParams['figure.figsize'] = [13, 5]
    plt.plot(HSI['data'])
    #Labling the plot
    plt.title('Buying House Related Searches')
    plt.ylabel('Search level changes %', fontsize=12)
    plt.legend(HSI['query'], loc='upper center', bbox_to_anchor=(0.5, -0.06),
          fancybox=True, shadow=True, ncol=4)

plt.show()
```

Plot below is the plot for the decomposition of the dataseries using PCA

```python
with plt.style.context('Solarize_Light2'):
    plt.rcParams['figure.figsize'] = [13, 5]
    plt.plot(HSI['index'], label = 'Search Index')    
    #Labling the plot
    plt.title('House Search index')
    plt.ylabel('Search level change %', fontsize=12)
    plt.legend(loc='upper left', bbox_to_anchor=(0.5, -0.06),
          fancybox=True, shadow=True, ncol=4)
    
plt.show()
```
